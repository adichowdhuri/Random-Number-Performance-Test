```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# [1, 2, 3, 4, 5]
# len(arr) = 5 // len(set(arr)) = 5
# [1, 1, 2, 3, 4]
# len(arr = 5) // len(set(arr)) = 4
```

```python
def distscore(arr):
    unq = set(arr)
    error = (len(arr)/len(unq))*100-100
    print(error,"%")
```

```python
def showPlots(arr):
    # Creating the hist plot
    plt.figure(figsize=(10, 6))
    plt.hist(arr)
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.title('Frequency of Each Number in the Array')
    plt.grid(True)
    plt.show()
    # Counting the frequency of each number in the array
    frequencies = {num: arr.count(num) for num in set(arr)}

    # Sorting the frequencies by number
    numbers = sorted(frequencies.keys())
    counts = [frequencies[number] for number in numbers]

    # Creating the line plot
    plt.figure(figsize=(10, 6))
    plt.plot(numbers, counts, marker='o', linestyle='-', color='blue')
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.title('Frequency of Each Number in the Array')
    plt.xticks(numbers)  # Set the x-ticks to match the numbers, if they are not too many
    plt.grid(True)
    plt.show()
```

```python
perf_arr = list(range(1, 11))
```

```python
bad_arr1 = [1]
bad_arr1 *=10
```

```python
rand_arr = [np.random.randint(1, 10) for _ in range(10)]
```
